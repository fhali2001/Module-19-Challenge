```python
# Import required libraries and dependencies
import pandas as pd
import hvplot.pandas
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Load the data into a Pandas DataFrame
df_market_data = pd.read_csv(
    "Resources/crypto_market_data.csv",
    index_col="coin_id")

# Display sample data
df_market_data.head(10)
```

```python
# Generate summary statistics
df_market_data.describe()
```

```python
# Plot your data to see what's in your DataFrame
df_market_data.hvplot.line(
    width=800,
    height=400,
    rot=90
)
```

---

### Prepare the Data

```python
from sklearn.preprocessing import StandardScaler
# Use the `StandardScaler()` module from scikit-learn to normalize the data from the CSV file
market_data_scaled = StandardScaler().fit_transform(df_market_data)
```

```python
# Create a DataFrame with the scaled data
df_market_data_scaled = pd.DataFrame(market_data_scaled, columns=df_market_data.columns)

# Copy the crypto names from the original data
crypto_names = df_market_data.index

# Set the 'coin_id' column as the index
df_market_data_scaled.index = crypto_names

# Display sample data
df_market_data_scaled.head(10)
```

---

### Find the Best Value for k Using the Original Data.

```python
# Create a list with the number of k-values from 1 to 10
k = list(range(1, 11))
inertia = []

# Compute the inertia with each possible value of k
for i in k:
    model = KMeans(n_clusters=i, random_state=0)
    model.fit(df_market_data_scaled)
    inertia.append(model.inertia_)

# Create a DataFrame with the data to plot the Elbow curve
df_elbow = pd.DataFrame({'k': k, 'inertia': inertia})

# Plot the Elbow curve
df_elbow.hvplot.line(x='k', y='inertia', title='Elbow Curve', xticks=k, xlabel='Number of Clusters (k)', ylabel='Inertia')
```

```python
# Create an empty list to store the inertia values
inertia = []

# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using `df_market_data_scaled`
# 3. Append the model.inertia_ to the inertia list
for i in k:
    model = KMeans(n_clusters=i, random_state=0)
    model.fit(df_market_data_scaled)
    inertia.append(model.inertia_)
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_data = {"k": k, "inertia": inertia}

# Create a DataFrame with the data to plot the Elbow curve
df_elbow = pd.DataFrame(elbow_data)
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.
df_elbow.hvplot.line(x="k", y="inertia", title="Elbow Curve", xticks=k, xlabel="Number of Clusters (k)", ylabel="Inertia")
```

### Cluster Cryptocurrencies with K-means Using the Original Data

```python
# Normalize the data
market_data_scaled = StandardScaler().fit_transform(df_market_data)

# Perform K-Means clustering on the scaled data
model = KMeans(n_clusters=4, random_state=0)
model.fit(market_data_scaled)
predictions = model.predict(market_data_scaled)

# Perform PCA on the scaled data
pca = PCA(n_components=3)
market_data_pca = pca.fit_transform(market_data_scaled)

# Perform K-Means clustering on the PCA data
model_pca = KMeans(n_clusters=4, random_state=0)
model_pca.fit(market_data_pca)
predictions_pca = model_pca.predict(market_data_pca)

# Create DataFrames for the results
df_market_data_predictions = pd.DataFrame(market_data_scaled, columns=df_market_data.columns)
df_market_data_predictions['cluster'] = predictions

df_market_data_pca_predictions = pd.DataFrame(market_data_pca, columns=["PC1", "PC2", "PC3"])
df_market_data_pca_predictions['cluster'] = predictions_pca

# Plot the results
df_market_data_predictions.hvplot.scatter(x="price_change_percentage_24h", y="price_change_percentage_7d", by="cluster", hover_cols=["coin_id"], title="Cryptocurrency Clusters", xlabel="Price Change Percentage (24h)", ylabel="Price Change Percentage (7d)")

df_market_data_pca_predictions.hvplot.scatter(x="PC1", y="PC2", by="cluster", hover_cols=["coin_id"], title="Cryptocurrency Clusters (PCA)", xlabel="Principal Component 1", ylabel="Principal Component 2")
```

### Visualize and Compare the Results

In this section, you will visually analyze the cluster analysis results by contrasting the outcome with and without using the optimization techniques.

```python
# Composite plot to contrast the clusters
cluster_plot_combined = (
    df_market_data_predictions.hvplot.scatter(
        x="price_change_percentage_24h",
        y="price_change_percentage_7d",
        by="cluster",
        hover_cols=["coin_id"],
        title="Cryptocurrency Clusters (Original Data)",
        xlabel="Price Change Percentage (24h)",
        ylabel="Price Change Percentage (7d)"
    ) +
    df_market_data_pca_predictions.hvplot.scatter(
        x="PC1",
        y="PC2",
        by="cluster",
        hover_cols=["coin_id"],
        title="Cryptocurrency Clusters (PCA Data)",
        xlabel="Principal Component 1",
        ylabel="Principal Component 2"
    )
).cols(2)
cluster_plot_combined
```
